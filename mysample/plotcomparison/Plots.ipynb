```python
import lammps_logfile
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
log1 = lammps_logfile.File("log1.lammps")
df1  = pd.DataFrame(log1.data_dict)
log2 = lammps_logfile.File("log2.lammps")
df2  = pd.DataFrame(log2.data_dict)
log3 = lammps_logfile.File("log3.lammps")
df3  = pd.DataFrame(log3.data_dict)
```

```python
vec_ones = np.ones(len(df1['PotEng'])) 

pe01 = df1['PotEng'][0]*vec_ones
ke01 = df1['KinEng'][0]*vec_ones
Eb01 = df1['E_bond'][0]*vec_ones
Ea01 = df1['E_angle'][0]*vec_ones
Ed01 = df1['E_dihed'][0]*vec_ones
Ei01 = df1['E_impro'][0]*vec_ones
Evdw01 = df1['E_vdwl'][0]*vec_ones 

pe1 = df1['PotEng']   - pe01
ke1 = df1['KinEng']   - ke01
Eb1 = df1['E_bond']   - Eb01
Ea1 = df1['E_angle']  - Ea01
Ed1 = df1['E_dihed']  - Ed01
Ei1 = df1['E_impro']  - Ei01
Evdw1 = df1['E_vdwl'] - Evdw01

pe02 = df2['PotEng'][0]*vec_ones
ke02 = df2['KinEng'][0]*vec_ones
Eb02 = df2['E_bond'][0]*vec_ones
Ea02 = df2['E_angle'][0]*vec_ones
Ed02 = df2['E_dihed'][0]*vec_ones
Ei02 = df2['E_impro'][0]*vec_ones
Evdw02 = df2['E_vdwl'][0]*vec_ones 

pe2 = df2['PotEng']   - pe02
ke2 = df2['KinEng']   - ke02
Eb2 = df2['E_bond']   - Eb02
Ea2 = df2['E_angle']  - Ea02
Ed2 = df2['E_dihed']  - Ed02
Ei2 = df2['E_impro']  - Ei02
Evdw2 = df2['E_vdwl'] - Evdw02

pe03 = df3['PotEng'][0]*vec_ones
ke03 = df3['KinEng'][0]*vec_ones
Eb03 = df3['E_bond'][0]*vec_ones
Ea03 = df3['E_angle'][0]*vec_ones
Ed03 = df3['E_dihed'][0]*vec_ones
Ei03 = df3['E_impro'][0]*vec_ones
Evdw03 = df3['E_vdwl'][0]*vec_ones 

pe3 = df3['PotEng']   - pe03
ke3 = df3['KinEng']   - ke03
Eb3 = df3['E_bond']   - Eb03
Ea3 = df3['E_angle']  - Ea03
Ed3 = df3['E_dihed']  - Ed03
Ei3 = df3['E_impro']  - Ei03
Evdw3 = df3['E_vdwl'] - Evdw03
```

```python
fig = px.line(df1,x='Time',y=[pe1, pe2, pe3, ke1, ke2, ke3])
fig.data[0].name = 'E_Pot 1x2'
fig.data[1].name = 'E_pot 5x10'
fig.data[2].name = 'E_pot 10x20'
fig.data[3].name = 'E_kin 1x2'
fig.data[4].name = 'E_kin 5x10'
fig.data[5].name = 'E_kin 10x20'
fig.update_yaxes(title="Energy Variation [kcal/mol]")
fig.update_xaxes(title="Time [fs]")
fig.show()
```

```python
fig = px.line(df1,x='Time',y=[Eb1, Eb2, Eb3])
fig.data[0].name = 'E_bond 1x2'
fig.data[1].name = 'E_bond 5x10'
fig.data[2].name = 'E_bond 10x20'
fig.update_yaxes(title="Energy Variation [kcal/mol]")
fig.update_xaxes(title="Time [fs]")
fig.show()
```

```python
fig = px.line(df1,x='Time',y=[Ea1, Ea2, Ea3, Ed1, Ed2, Ed3, Ei1, Ei2, Ei3])
fig.data[0].name = 'E_angle 1x2'
fig.data[1].name = 'E_angle 5x10'
fig.data[2].name = 'E_angle 10x20'
fig.data[3].name = 'E_dihed 1x2'
fig.data[4].name = 'E_dihed 5x10'
fig.data[5].name = 'E_dihed 10x20'
fig.data[6].name = 'E_impro 1x2'
fig.data[7].name = 'E_impro 5x10'
fig.data[8].name = 'E_impro 10x20'
fig.update_yaxes(title="Energy Variation [kcal/mol]")
fig.update_xaxes(title="Time [fs]")
fig.show()
```

```python
fig = px.line(df1,x='Time',y=[Evdw1, Evdw2, Evdw3])
fig.data[0].name = 'E_vdwl 1x2'
fig.data[1].name = 'E_vdwl 5x10'
fig.data[2].name = 'E_vdwl 10x20'
fig.update_yaxes(title="Energy Variation [kcal/mol]")
fig.update_xaxes(title="Time [fs]")
fig.show()
```
